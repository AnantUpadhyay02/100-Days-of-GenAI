#Eigendecomposition
=====================================
Affine transformation via Matrix Application

```python
vec = np.array([2,4])
I = np.array([[1,0],[0,1]]) # Identity Matrix, does nothing to the vector
E = np.array([[1,0],[0,-1]])# It flips the matrix on the x axis
Evec = np.dot(E,vec)
Evec
```

```python
from matplotlib import pyplot as plt
def plot_vectors(vectors, colors):
    plt.figure()
    plt.axvline(x = 0,color= 'lightgrey')
    plt.axhline(y = 0, color = 'lightgrey')
    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]],[x[1]],[x[2]],[x[3]],angles = 'xy',scale_units = 'xy', scale = 1, color= colors[i])

```

```python
plot_vectors([vec,Evec],['orange','red'])
plt.xlim(-1,5)
_ = plt.ylim(-10,10)
```

Applying a flipping matrix is and example of affine transformation i.e. a change in geometry that mar adjust distances or angles between the vectors,but preserve parallelism between them
