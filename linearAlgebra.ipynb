#Eigendecomposition
=====================================
Affine transformation via Matrix Application

```python
vec = np.array([2,4])
I = np.array([[1,0],[0,1]]) # Identity Matrix, does nothing to the vector
E = np.array([[1,0],[0,-1]])# It flips the matrix on the x axis
Evec = np.dot(E,vec)
Evec
```

```python
from matplotlib import pyplot as plt
def plot_vectors(vectors, colors):
    plt.figure()
    plt.axvline(x = 0,color= 'lightgrey')
    plt.axhline(y = 0, color = 'lightgrey')
    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]],[x[1]],[x[2]],[x[3]],angles = 'xy',scale_units = 'xy', scale = 1, color= colors[i])

```

```python
plot_vectors([vec,Evec],['orange','red'])
plt.xlim(-1,5)
_ = plt.ylim(-10,10)
```

Applying a flipping matrix is and example of affine transformation i.e. a change in geometry that mar adjust distances or angles between the vectors,but preserve parallelism between them

# SVD: Single Value Decomposition
For a matrix A: 
A = U D V(transpose)
====================

where, U is the orthogonal m X m matix of left singular Vectors
    D is the diagonal m X n matrix, elements along it's diagonal are singular values of matrix A
    V is the orthogonal n X n matrix of right singular Vectors of A


```python
a = np.array([[1,2],[-3,4],[5,6]])
a
```

```python
u,d,vt = np.linalg.svd(a)
u,d,vt
```

# Image Compression using SVD

```python
from PIL import Image
```

```python
img = Image.open('556319.jpg')
_ = plt.imshow(img)
```

```python
# To avoid complexity of multiple color channels we can use greyscale 
imgrey = img.convert('LA')
_ = plt.imshow(imgrey)
```

```python
if isinstance(imgrey, Image.Image):
    print("imgrey is a PIL image")
else:
    print("imgrey is not a PIL image")
```

```python
# Converting Image into the matrix data
imgMat = np.array(list(imgrey.getdata(band=0)),float)
imgMat.shape = (imgrey.size[1],imgrey.size[0])
imgMat = np.matrix(imgMat)
_ = plt.imshow(imgMat,cmap = 'gray')
```

```python
U , sigma , v = np.linalg.svd(imgMat)
U, sigma ,v
```

```python
# this is to compress the image using SVD
for i in [2, 4, 8, 16, 32, 64]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(v[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

The image represented as 64 singular vectors is significantly of the smaller size of the original image

# Moore - Penrose Pseudoinversion

A(+) = V D(+) U(transpose)

```python
a
u, d, vt = np.linalg.svd(a)
u,d,vt
```

```python
#here we have to create D(+)
D = np.diag(d)
Dinv = np.linalg.inv(D)
Dinv

#Dplus must have dimensions of A transpose for matrix multiplication to be possible

Dplus = np.concatenate((Dinv,np.array([[0,0]]).T),axis = 1)
Dplus
```

```python
Aplus = np.dot(vt.T,np.dot(Dplus,u.T))
Aplus
#here this is the pseudoinverse of matrix A
```

```python
#Altho numpy already have an attribute for pseudoinverse
Aplus = np.linalg.pinv(a)
Aplus
```
