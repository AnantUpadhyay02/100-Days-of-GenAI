```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0,50,10000)
d_r = 2.5*t
d_s = 3*(t-5)
```

```python
fig, ax = plt.subplots()
plt.title("The Rober Chase")
plt.xlabel("Time in mins")
plt.ylabel("Distance in km")
ax.set_xlim([0,60])
ax.set_ylim([0,100])
ax.plot(t,d_r,c = "red")
ax.plot(t,d_s, c= "green")
```
