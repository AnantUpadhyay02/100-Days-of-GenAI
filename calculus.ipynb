Limits

```python
# The calculus of infinitism
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-10,10,1000)# this is to set the limit of x for the equation
y =x**2 + 2*x + 2
```

```python
fig , ax = plt.subplots()
_=ax.plot(x,y)
```

```python
fig , ax = plt.subplots()
ax.set_xlim(-2,0)
ax.set_ylim(0,2)
_ = ax.plot(x,y)
```

# #Day 9 
The indeterminant 0/0 cases handled here, can always use L'hospital rule in the Pen Paper solution

```python
def my_func(eq_x):
    eq_y = (eq_x**2 - 1)/(eq_x-1)
    return eq_y
```

```python
my_func(0.99999),my_func(1.0001)
#the closer and closer we get to one we are getting close to 2, but at 1 we can't divide
```

```python
y = my_func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axvline(x=1, color='purple', linestyle='--')
plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

```python
def trig_func(eq_x):
    eq_y = np.sin(eq_x)/eq_x
    return eq_y
```

```python
y = trig_func(x)
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-1, 2)
plt.axvline(x=0, color='purple', linestyle='--')
plt.axhline(y=1, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

```python
def infi_func(eq_x):
    eq_y = 30/eq_x
    return eq_y
```

```python
y = infi_func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
# plt.axvline(x=1, color='purple', linestyle='--')
# plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x,y)

y= infi_func(x)
```
