# Calculus

Limits
============

```python
# The calculus of infinitism
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-10,10,1000)# this is to set the limit of x for the equation
y =x**2 + 2*x + 2
```

```python
fig , ax = plt.subplots()
_=ax.plot(x,y)
```

```python
fig , ax = plt.subplots()
ax.set_xlim(-2,0)
ax.set_ylim(0,2)
_ = ax.plot(x,y)
```
