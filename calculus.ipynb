# #Day 9 
The indeterminant 0/0 cases handled here, can always use L'hospital rule in the Pen Paper solution

```python
def my_func(eq_x):
    eq_y = (eq_x**2 - 1)/(eq_x-1)
    return eq_y
```

```python
my_func(0.99999),my_func(1.0001)
#the closer and closer we get to one we are getting close to 2, but at 1 we can't divide
```

```python
y = my_func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axvline(x=1, color='purple', linestyle='--')
plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

```python
def trig_func(eq_x):
    eq_y = np.sin(eq_x)/eq_x
    return eq_y
```

```python
y = trig_func(x)
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-1, 2)
plt.axvline(x=0, color='purple', linestyle='--')
plt.axhline(y=1, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

```python
def infi_func(eq_x):
    eq_y = 30/eq_x
    return eq_y
```

```python
y = infi_func(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-300, 300)
# plt.axvline(x=1, color='purple', linestyle='--')
# plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x,y)

y= infi_func(x)
```

# #Day10
Differentiation and Derivative computation

```python
def f(eq_x):
    eq_y = eq_x**2+eq_x*2+2
    return eq_y
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x = 0, color = "grey")
plt.axhline(y = 0, color = "grey")
_ = plt.plot(x,y)
```

Find the slope at a point, say at x = 2

```python
f(2),f(5) #that means y = 10 at x = 2
```

```python
fig, ax = plt.subplots()
plt.axvline(x = 0, color = "grey")
plt.axhline(y = 0, color = "grey")
plt.scatter(2,10)
plt.scatter(5,37,c = "red",zorder = 3) # this zorder attribute is to keep the point above the curve
_ = plt.plot(x,y)
```

Now to find the slope between the two points using the standard mathematical formula: (y2-y1)/(x2-x1)


```python
m = (37-10)/(5-2)
m
```

now plotting the line with the slope we just calculated : y = mx + b

```python
b = 37-m*5
b
```

```python
line_y = m*x+b
```

```python

plt.fig, ax = plt.subplots()
plt.axvline(x = 0, color = "grey")
plt.axhline(y = 0, color = "grey")
plt.scatter(5,37,c="red",zorder = 3)
plt.scatter(2,10)
plt.ylim(-5,150)
plt.plot(x,line_y,c= "orange",)
_ = plt.plot(x,y)
```

```python
#we gonna use pytorch and tensorflow for automatic differentiation
import torch
```

```python
x= torch.tensor(10.0)
x
```

```python
x.requires_grad_() #constantly tracks gradients through forward pass
```

```python
y =  x**2
```

```python
y.backward() #autodiff
y
```

```python
x.grad #x became 20
```
